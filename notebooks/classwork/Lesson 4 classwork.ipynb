```python
#calculate the slugging average, and get the score from the best
```

```python
slg = lambda x: (x['h']-x['X2b']-x['X3b']-x['hr'] + 2*x['X2b'] + 3*x['X3b'] + 4*x['hr'])/(x['ab']+1e-6)
basebal_stats = baseball[['h', 'X2b', 'X3b', 'hr', 'ab']]
baseball['SLG'] = basebal_stats.apply(slg, axis=1)
baseball['SLG'] - baseball['SLG'].max()
```
